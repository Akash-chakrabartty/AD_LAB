```python
import matplotlib.pyplot as plt
import numpy as np

# .   (Table 6  )
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Oct', 'Nov', 'Dec']
actual = [112, 113, 114, 115, 112, 121, 122, 114]
predicted = [113, 112, 116, 117, 110, 118, 121, 115]

def solve_step_by_step():
    n = len(actual)
    total_abs_error = 0
    total_squared_error = 0
    
    #     
    print(f"{'Month':<10} | {'Actual':<8} | {'Predicted':<10} | {'Abs Error':<10} | {'Sq Error':<10}")
    print("-" * 65)
    
    #      
    for i in range(n):
        diff = actual[i] - predicted[i]
        abs_err = abs(diff) #  
        sq_err = diff ** 2  #  
        
        total_abs_error += abs_err
        total_squared_error += sq_err
        
        #    
        print(f"{months[i]:<10} | {actual[i]:<8} | {predicted[i]:<10} | {abs_err:<10} | {sq_err:<10}")
    
    print("-" * 65)
    
    #   
    mae = total_abs_error / n
    mse = total_squared_error / n
    
    print(f"Final MAE (13 / 8) = {mae:.3f}")
    print(f"Final MSE (25 / 8) = {mse:.3f}")

    # .   (Essential Graph)
    plt.figure(figsize=(10, 6))
    plt.plot(months, actual, marker='o', label='Actual Sales', color='blue')
    plt.plot(months, predicted, marker='x', label='Predicted Sales', color='red', linestyle='--')
    plt.title('Performance Comparison: Actual vs Predicted Sales')
    plt.xlabel('Months')
    plt.ylabel('Sales in INR Lakhs')
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.show()

#   
solve_step_by_step()
```
