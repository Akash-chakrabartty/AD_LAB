```python
import numpy as np
import matplotlib.pyplot as plt

# .   (Table 7  )
x = np.array([1, 1, 2, 3, 3, 5, 5, 6]) # No. of competitors
y = np.array([3600, 3300, 3100, 2900, 2700, 2300, 2000, 1800]) # Sales

def medical_store_regression():
    n = len(x)
    
    # .      (Manual Step Emulation)
    sum_x = 0
    sum_y = 0
    sum_xy = 0
    sum_x_sq = 0
    
    print(f"{'X':<5} | {'Y':<8} | {'X^2':<8} | {'XY':<10}")
    print("-" * 40)
    
    for i in range(n):
        x_sq = x[i]**2
        xy = x[i] * y[i]
        
        sum_x += x[i]
        sum_y += y[i]
        sum_xy += xy
        sum_x_sq += x_sq
        
        print(f"{x[i]:<5} | {y[i]:<8} | {x_sq:<8} | {xy:<10}")
    
    print("-" * 40)
    
    # .   (Slope m  Intercept c)  
    # : m = (n*sum_xy - sum_x*sum_y) / (n*sum_x_sq - sum_x^2)
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_sq - sum_x**2)
    
    # : c = (sum_y - m*sum_x) / n
    c = (sum_y - m * sum_x) / n
    
    print(f"Total Sum X: {sum_x}")
    print(f"Total Sum Y: {sum_y}")
    print(f"Slope (m): {m:.2f}")
    print(f"Intercept (c): {c:.2f}")
    print(f"Equation: Sales = {m:.2f} * Competitors + {c:.2f}")

    # .  : Scatter Diagram (  )
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='blue', label='Actual Sales Data')
    plt.title('Scatter Diagram: No. of Competitors vs Sales')
    plt.xlabel('Number of Competitors (within 1km)')
    plt.ylabel('Daily Sales Volume')
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.legend()
    plt.show()

    # .  : Linear Regression Model (  )
    plt.figure(figsize=(10, 5))
    plt.scatter(x, y, color='blue', label='Data Points')
    y_pred = m * x + c #  
    plt.plot(x, y_pred, color='red', linewidth=2, label='Linear Regression Line')
    plt.title('Regression Model: Sales Volume Prediction')
    plt.xlabel('Number of Competitors')
    plt.ylabel('Sales Volume')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    medical_store_regression()
```
