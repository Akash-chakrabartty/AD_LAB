```python
import matplotlib.pyplot as plt

def solve_lab_assignment():
    # ১. ডেটা ইনপুট (Data Input)
    # x = Age (Independent Variable), y = Glucose Level (Dependent Variable)
    x = [43, 21, 25, 42, 57, 59]
    y = [99, 65, 79, 75, 87, 81]
    n = len(x)
    target_age = 49

    # ২. ক্যালকুলেশন ভেরিয়েবল (Initialization)
    sum_x = 0
    sum_y = 0
    sum_x2 = 0
    sum_xy = 0

    # ৩. লুপ ব্যবহার করে টেবিলের মানগুলো বের করা (Using loops for calculation)
    print(f"{'Subject':<10} | {'Age(x)':<10} | {'Glucose(y)':<12} | {'x^2':<10} | {'xy':<10}")
    print("-" * 60)
    
    for i in range(n):
        sum_x += x[i]
        sum_y += y[i]
        sum_x2 += x[i]**2
        sum_xy += x[i] * y[i]
        # প্রতিটি লাইনের হিসাব প্রিন্ট করা
        print(f"{i+1:<10} | {x[i]:<10} | {y[i]:<12} | {x[i]**2:<10} | {x[i]*y[i]:<10}")

    print("-" * 60)
    print(f"Total Sums: | Σx={sum_x:<7} | Σy={sum_y:<10} | Σx2={sum_x2:<8} | Σxy={sum_xy:<8}")

    # ৪. Slope (m) এবং Intercept (c) বের করা
    # Formula: m = (n*Σxy - Σx*Σy) / (n*Σx2 - (Σx)^2)
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x**2)
    
    # Formula: c = (Σy - m*Σx) / n
    c = (sum_y - m * sum_x) / n

    # ৫. প্রেডিকশন (Prediction for Age 49)
    predicted_y = (m * target_age) + c

    print(f"\n--- RESULT---")
    print(f"Slope (m) = {m:.4f}")
    print(f"Intercept (c) = {c:.4f}")
    print(f"Regression Equation: y = {m:.4f}x + {c:.4f}")
    print(f"Predicted Glucose for age {target_age} = {predicted_y:.2f}")

    # ৬. গ্রাফ তৈরি (Generating Graph)
    plt.figure(figsize=(10, 6))
    
    # আসল ডেটা পয়েন্টগুলো (Blue dots)
    plt.scatter(x, y, color='blue', label='Actual Data Points')
    
    # রিগ্রেশন লাইন (Red Line)
    line_x = [min(x)-5, max(x)+5]
    line_y = [m * val + c for val in line_x]
    plt.plot(line_x, line_y, color='red', linewidth=2, label='Regression Line (Best Fit)')
    
    # প্রেডিকশন পয়েন্ট (Green star)
    plt.scatter(target_age, predicted_y, color='green', marker='*', s=200, label=f'Predicted Value for age {target_age}')

    # গ্রাফের সাজসজ্জা
    plt.title('Age vs Glucose Level: Linear Regression Analysis', fontsize=14)
    plt.xlabel('Age (x)', fontsize=12)
    plt.ylabel('Glucose Level (y)', fontsize=12)
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    
    plt.show()

# কোডটি রান করা
solve_lab_assignment()
```
