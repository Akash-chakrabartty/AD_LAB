```python
import numpy as np
import matplotlib.pyplot as plt

# .   (Table 7    )
# x =   (Number of Competitors)
# y =   (Daily Sales Volume)
competitors = np.array([1, 2, 3, 4, 5, 6, 7, 8])
sales = np.array([120, 110, 100, 90, 80, 75, 60, 50])

def develop_linear_model():
    n = len(competitors)
    
    # .   (Step-by-step Summations)
    sum_x = np.sum(competitors)
    sum_y = np.sum(sales)
    sum_xy = np.sum(competitors * sales)
    sum_x_sq = np.sum(competitors**2)
    
    #  (Slope, m) 
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_sq - sum_x**2)
    
    #  (Intercept, c) 
    c = (sum_y - m * sum_x) / n

    print("--- Linear Regression Calculation ---")
    print(f"Total Sum of X (Competitors): {sum_x}")
    print(f"Total Sum of Y (Sales): {sum_y}")
    print(f"Calculated Slope (m): {m:.4f}")
    print(f"Calculated Intercept (c): {c:.4f}")
    print("-" * 40)
    print(f"Final Model Equation: Sales = {m:.4f} * (Competitors) + {c:.4f}")

    # .  : Scatter Diagram
    plt.figure(figsize=(10, 5))
    plt.scatter(competitors, sales, color='blue', label='Actual Sales Data')
    plt.title('Scatter Diagram: Sales vs Competitors')
    plt.xlabel('Number of Competitors (within 1km)')
    plt.ylabel('Daily Sales Volume')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    # .  : Regression Line (Model Visualization)
    plt.figure(figsize=(10, 5))
    plt.scatter(competitors, sales, color='blue', label='Data Points')
    #   
    y_pred = m * competitors + c
    plt.plot(competitors, y_pred, color='red', linewidth=2, label='Regression Line (Linear Model)')
    plt.title('Linear Regression Model for Medical Stores')
    plt.xlabel('Number of Competitors')
    plt.ylabel('Sales Volume')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

if __name__ == "__main__":
    develop_linear_model()
```
