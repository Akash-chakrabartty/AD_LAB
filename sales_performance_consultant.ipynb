```python
import numpy as np
import matplotlib.pyplot as plt

# .   (Table 6  )
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Oct', 'Nov', 'Dec']
actual = [112, 113, 114, 115, 112, 121, 122, 114]
predicted = [113, 112, 116, 117, 110, 118, 121, 115]

def run_performance_consultancy():
    n = len(actual)
    total_sq_err = 0
    total_abs_pct_err = 0
    
    #      
    print(f"--- Step-by-Step Sales Analysis Report ---")
    print(f"{'Month':<8} | {'Actual':<8} | {'Predicted':<10} | {'Sq Error':<10} | {'Abs % Error'}")
    print("-" * 65)
    
    # .     (Detailed Steps)
    for i in range(n):
        #  
        diff = actual[i] - predicted[i]
        sq_err = diff ** 2
        abs_pct_err = abs(diff) / actual[i]
        
        total_sq_err += sq_err
        total_abs_pct_err += abs_pct_err
        
        #    
        print(f"{months[i]:<8} | {actual[i]:<8} | {predicted[i]:<10} | {sq_err:<10} | {abs_pct_err:.5f}")
    
    print("-" * 65)
    
    # .   
    mse = total_sq_err / n
    rmse = np.sqrt(mse)
    
    #   : Hybrid Error = 0.3 * MSE + 0.25 * RMSE
    hybrid_error = (0.3 * mse) + (0.25 * rmse)
    
    # MAPE 
    mape = (total_abs_pct_err / n) * 100

    print(f"Total Squared Error (Sum): {total_sq_err}")
    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"--- Calculated Hybrid Error: {hybrid_error:.4f} ---")
    print(f"--- Mean Absolute Percentage Error (MAPE): {mape:.4f}% ---")

    # .   : Actual vs Predicted (Line Graph)
    plt.figure(figsize=(10, 5))
    plt.plot(months, actual, marker='o', label='Actual Sales', color='blue', linewidth=2)
    plt.plot(months, predicted, marker='x', linestyle='--', label='Predicted Sales', color='red', linewidth=2)
    plt.title('Sales Comparison: Actual vs Predicted (B2C Firm)')
    plt.xlabel('Months')
    plt.ylabel('Sales (INR Lakhs)')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

    # .   : Monthly Percentage Error (Bar Chart)
    #  MAPE   
    plt.figure(figsize=(10, 5))
    monthly_mape = [(abs(actual[i] - predicted[i]) / actual[i]) * 100 for i in range(n)]
    plt.bar(months, monthly_mape, color='teal', alpha=0.7, label='Monthly Error %')
    plt.axhline(y=mape, color='orange', linestyle='-', label=f'Avg MAPE ({mape:.2f}%)')
    plt.title('Error Analysis: Monthly MAPE (%)')
    plt.xlabel('Month')
    plt.ylabel('Error Percentage (%)')
    plt.legend()
    plt.show()

#   
if __name__ == "__main__":
    run_performance_consultancy()
```
